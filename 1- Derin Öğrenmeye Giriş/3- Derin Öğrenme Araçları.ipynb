# Derin Öğrenme Araçları

- ***TensorFlow:*** 2015 yılında Google tarafından geliştirilen açık kaynak kodlu bir derin öğrenme kütüphanesidir. Esnek yapısı sayesinde bir ya da birden fazla CPU ya da GPU kullanarak projeleri geliştirmeye olanak sağlar. Bu kadar bilinmesi ve yaygınlaşmasının sebeplerinden bir Google, diğeri ise python'dır çünkü TensorFlow python ile yazılmış ve python ile kullanılmaktadır.

- ***Keras:*** Python'da yazılmış açık kaynak kodlu bir derin öğrenme kütüphanesidir. İçerdiği çok fazla işlevsel fonksiyon sayesinde kolayca bir derin öğrenme modeli oluşturup eğitmemizi sağlar. 2017 yılında Google'ın TensorFlow ekibi TensorFlow'un çekirdek kütüphanesinde Keras'ı desteklemeye karar vermiştir. TensorFlow'a daha basit arayüzler geliştirecek şekilde konumlanmıştır. Daha az çabayla daha verimli işler ortaya çıkarılmasını sağlamaktadır.

- ***PyTorch:*** 2017 yılında Facebook tarafından Torch'a bağlı olan python tabanlı ve açık kaynak kodlu makine öğrenmesi kütüphanesi olarak yayınlanmıştır. Projelerin içerisinde GPU'ları kullanarak yapısı gereği sağladığı esneklik ve hız ile günümüzde oldukça yaygın kullanılmaktadır. TensorFlow mu PyTorch mu şeklinde bir atışma da söz konusudur.

- ***Scikit Learn:*** Tam olarak GPU desteği sağlamasa da derin sinir ağı geliştirmek için birçok araç sağlamaktadır. 

- **OpenCV (Open Source Computer Vision):**  Gerçek zamanlı görüntü işleme uygulamalarında kullanılan açık kaynak kodlu bir kütüphanedir. Derin öğrenme dediğimizde görüntü, ses, metin gibi çeşitli veri yapısı üzerinde çalışma imkanımız olacaktır. Fakat derin öğrenmenin kendisiyle en çok bütünleşen uygulamalardan birisi görüntü işlemedir. 

- **Caffe:** Diğer araçlara benzer şekilde derin öğrenme için kullanılabilen, hızlı, modüler bir derin öğrenme framework'üdür. 

- **Knet.jl:** Koç Üniversitesi'nden Deniz Yüret tarafından geliştirilmiş bir kütüphanedir. Julia programlama dilinde yazılmıştır.


Biz bu ders kapsamında TensorFlow'u kullanarak işlemlerimiz gerçekleştireceğiz.
