# Kısa Tarihçe

- 1950: Turing bu alandaki en önemli isimlerden biridir. Özellikle yapay zeka ile ilgili henüz kendi zamanında birçok gelişme olmamasına rağmen çok önemli katkıları olmuştur. İleride bilgisayarların da insanlar gibi düşünebileceğini hatta insanları da geçeceğini söylemiştir.

- 1955: John McCarthy, Turing testini geçecek bilgisayarlar için uygun yazılımların yazılabileceğine inanıyordu. Bunu geliştirebilmek amacıyla "yapay zeka" kavramını ortaya atmış ve bu sayede "yapay zeka" bir bilim dalı olarak gelişmiştir.

- 1957: Amerikalı psikolog Frank Rosenblatt, öğrenme süreci sırasında beyin nöronlarının uyarılmasından esinlenerek "Perceptron (Algılayıcı)" kavramını ortaya çıkarmıştır. 

- 1967: Marvin Minsky, otomatik öğrenme işini kendini güncelleyerek her adımda ilerlenebilme durumunu ifade etmiştir. Asimov, Minsky hakkında şunları söylemiştir: "Benden zeki olduğunu kabul ettiğim iki kişiden birisidir". Minsky henüz bilgisayar kavramının yeni yeni çıktığı dönemde dahi yaptığı araştırmalarla bilgisayar biliminin ilerlemesinde çok ciddi roller almıştır. 1959 yılında MIT'nin yapay zeka laboratuvarını kuran kişidir.

- 1997: Deep Blue, IBM tarafından geliştirilmiş, satranç oynayabilen bir bilgisayardır. Bu bilgisayar ilk satranç oyununda Kasparov ile 1-1 berabere kalmıştır. Ancak sonrasında IBM tarafından geliştirilerek saniyede 200 milyon pozisyon deneyebilecek hale getirilmiş ve 1997'de dünya satranç şampiyonu Kasparov'u yenmeyi başarmıştır.

- 2006: Derin Yapay Sinir Ağı, sinir ağının çalışma prensibini derinleştirerek (birden fazka gizli katman kullanarak) öğrenme işlemlerini gerçekleştirmeye başlamasıyla derin öğrenme alanındaki çalışmalar çok daha farklı bir boyuta gelmiştir. Hatta öyleki günümzde derin öğrenme dendiğinde bütün sinir ağları artık derindir. Dolayısıyla artık daha derin mimariler ile bakış kazanmamızı sağlayan önemli bir dönüm noktasıdır.

- 2011: Artık sesli asistanlar çeşitli sorulara yanıt verebilmekte, doğal dil ile sorulan sorulara cevap verebilmekte, tavsiyelerde bulunabilmekte ve çeşitli web hizmetleri için destek verebilmektedir.

- 2012: Grafik işlemcileri bu tarihte öne çıkaran "Large Scale Visual Recognition Challenge (ILSVRC)" isimli bir yarışmada görüntü işleme alanında en iyi hata oranını yarı yarıya indirerek birinci olan gruptur. Bunu yaparken de kullandıkları en önemli parça grafik işlemcileri olmuştur. Böylece yapay zeka için grafik işlemcileri de devreye girmiştir.

- 2014: Alexa, yaratıcıların algoritmalarını insan düşünce sürecini taklit ederek yapılandırılmış ve sesli asistanlar yapay zeka kavramını birkaç adım daha ileriye taşımıştır. Yine bu yulda GAN'lar hayatımıza girmiştri. GAN (Generative Adversarial Networks), belirli bir veri seti üzerinden öğrenerek yeni ve var olmayan insan yüzleri yaratılması veya belirli resimler incelenerek bu resimlere benzer resimler yaratılmasıdır.

- 2016: AlphaGO, Google Deep Mind tarafından geliştirilen ve GO oyununu oynayan bir programdır. Ekim 2015'te 19'a 19'luk tahtada profesyonel bir GO oyuncusunu avantaj verilmeden yenen ilk bilgisayar programı olmuştur.

- 2018: BERT, doğal dil işleme ve makine öğrenmesi alanında çok önemli bir tekniği oraya çıkarmıştır. Metinler üzerinden öğrenme işlemi gerçekleştirirken metinlerdeki anlamsal bilgileri tutup buna göre çıkarımlarda bulunarak çok önemli bir katkı sağlamıştır.

- 2019: GPT2 yönteminin geliştirilmesinde 8 milyon web sayfasından 40 GB'lık veri toplanmıştır. Yazılımın ilk modeli GPT için sadece 4 GB'lık bir veri kullanılmışken GPT2'de 40 GB'lık veri toplanması sayesinde yazılım bir süre sonra kendi kendine öğrenmeye (az bir detayla bile çok fazla çıkarımda bulunma) başlamıştır. Örneğin sadece ilk 2 cümlenin verilmesiye 9 paragraflık bir haber metnini yazabilmektedir. BERT ile birlikte doğal dil işleme alanında çok önemli dönüm noktalarıdır.

- 2020: GPT3, insanların yazdığı metinlere benzer içerik üretmek için derin öğrenmeyi kullanabilen özbağlanımlı bir dil modelidir. Toplanılan 570 GB'lık metin verileriyle bu model beslenmiştir.

- 2021: DALL-E, yine OpenAI tarafından üretilen metinsel tanımlardan görüntü oluşyuran bir yapay zeka programıdır. İsmini Salvador Dali ve WALL-E isimli animasyon robottan almıştır.
